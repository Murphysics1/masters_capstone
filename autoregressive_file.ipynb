```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
df_import = pd.read_csv('coffee-shop-sales-revenue.csv', 
                        delimiter= '|',
                        parse_dates=['transaction_date'],
                        index_col='transaction_id'
                        )

df_import.head()
```

```python
categories = ['Coffee','Tea']

df_filtered = df_import[df_import['product_category'].isin(categories)]
```

```python
df = df_filtered.groupby(['transaction_date','product_category'])['transaction_qty'].sum().reset_index()
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(12,5))

sns.lineplot(data=df,x='transaction_date',y='transaction_qty',hue='product_category',ax=ax)
ax.set_xlabel('Date')
ax.set_ylabel('Purchase Quantity')
ax.set_title('Coffee and Tea Purchases')
ax.legend(labels=)
plt.show()
```
